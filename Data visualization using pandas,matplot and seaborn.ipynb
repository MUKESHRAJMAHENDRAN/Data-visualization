```python
years = range(2000, 2012)
apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.9375, 0.9372, 0.939]
oranges = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898, 0.9, 0.896, ]
```

```python
plt.plot(years,apples);
plt.plot(years,oranges);
plt.xlabel("year")
plt.ylabel('yeild(tons per hectare)')
```

```python
plt.figure(figsize=(12,6));
plt.plot(years,apples,'b-o');
plt.plot(years,oranges,'g--x');
plt.xlabel("Years")
plt.ylabel('Yield(tons per hectare)')
plt.title('Agriculture production of Apple and oranges')
plt.legend(['Apples','Oranges']);
```

## Scatter plot

```python
flower_df=sns.load_dataset('iris')
```

```python
flower_df
```

```python
flower_df.species.unique()
```

```python
plt.plot(flower_df.sepal_length,flower_df.sepal_width)
```

To avoid this type graph we are using Scatter plot

```python
sns.scatterplot(x='sepal_length',y='sepal_width',hue=flower_df.species,data=flower_df,s=100)
```

```python
plt.figure(figsize=(12,7))
plt.title('Sepal dimension')
sns.scatterplot(x='sepal_length',y='sepal_width',
                hue=flower_df.species,
                data=flower_df,s=100);
```

# Histogtram

```python
flower_df.sepal_width
```

```python
plt.hist(flower_df.sepal_width)
```

```python
plt.hist(flower_df.sepal_width,bins=4);
```

```python
plt.hist(flower_df.sepal_width,bins=np.arange(2,5,0.25));
```

```python
setosa_df=flower_df[flower_df.species=="setosa"]
versicolor_df=[flower_df.species=="versicolor"]
virginica_df=flower_df[flower_df.species=="virginica"]
```

```python
plt.hist(setosa_df.sepal_width,alpha=0.4,bins=np.arange(2,5,0.25))
plt.hist(versicolor_df.sepal_width, alpha=0.4, bins=np.arange(2, 5, 0.25));
plt.hist(virginica_df.sepal_width, alpha=0.4, bins=np.arange(2, 5, 0.25));
```
